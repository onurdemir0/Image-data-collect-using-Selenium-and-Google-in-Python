```python
from six import BytesIO
from PIL import Image
import numpy as np
import glob
import os

import matplotlib
import matplotlib.pyplot as plt
```

```python
def load_image_into_numpy_array(path):
    img_data = open(path,'rb').read()
    image = Image.open(BytesIO(img_data))
    (im_width, im_height) = image.size
    return np.array(image.getdata()).reshape((im_height, im_width, 3)).astype(np.uint8)
```

```python
apple_image_path = "\Proje\images\fresh banana"
apple_images_np = []
for iname in glob.glob(apple_image_path):
    apple_images_np.append(load_image_into_numpy_array(iname))
```

```python
plt.rcParams['figure.figsize'] = [14,7]
for idx, apple_image_np in  enumerate(apple_images_np):
    plt.subplot(2, 3, idx+1)
    plt.imshow(apple_image_np)
plt.show
```
